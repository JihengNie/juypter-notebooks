```python
import pandas as pd
import numpy as np
```

```python
path = r'C:\Users\Gooogl3\MarketResearch.xlsx'
df = pd.read_excel(path)
```

```python
# Headers
headers = df.columns.values.tolist()
headers.pop(0)
headers.pop(0)
headers
```

```python
# Rows
rows = df['Column Names']
rows = rows.to_numpy().tolist()
rows.pop(0)
rows
```

# Overall

```python
df
```

# Percentage

```python
df1 = df.copy()
df1 = df1.iloc[1: , :]
del df1['Overall']
df1 = df1.reset_index(drop=True)
df1
```

```python
storageArray1 = []
for items in headers:
    tempObject = {'Category': items,
        'max_value': df1[items].max(),
        'max_index': df1[items].idxmax(),
        'row_value': rows[df1[items].idxmax()],
        }
    storageArray1.append(tempObject)
# storageArray1
```

```python
for items in storageArray1:
    print(items['Category']+
          '-- the largest percentage is '+ items['row_value'] + ' at '+ str(items['max_value'])+'%')
```

# Index

```python
df2 = df.copy()
df2 = df2.iloc[1: , :]
for items in headers:
    df2[items+' index'] = df2[items] / df2['Overall'] * 100
    del df2[items]
del df2['Overall']
df2 = df2.round(0)
df2 = df2.reset_index(drop=True)
df2
```

```python
storageArray2 = []
for items in headers:
    tempObject = {'Category': items+' index',
        'max_value': df2[items+' index'].max(),
        'max_index': df2[items+' index'].idxmax(),
        'row_value': rows[df2[items+' index'].idxmax()],
        }
    storageArray2.append(tempObject)
# storageArray2
```

```python
for items in storageArray2:
    print(items['Category']+
          '-- the largest index is '+ items['row_value'] + ' at '+ str(items['max_value']))
```

# Range

```python
df3 = df.copy()
del df3['Overall']
df3 = df3.iloc[1: , :]
df3['Max'] = df3.max(axis=1)
df3['Min'] = df3.min(axis=1)
df3['Range'] = df3['Max'] - df3['Min']
```

```python
df3[['Column Names', 'Range']].sort_values(
by = 'Range',
ascending = False)
```
